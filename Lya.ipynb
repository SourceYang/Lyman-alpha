```python
#plot the distribution of generated positions for Lyman alpha photons
import matplotlib.pyplot as plt

data_position = []

#the open function will be modified depend on your own directory
with open("D://Data/Lyp.txt") as f:
    for line in f:
        data = line.split()
        for x in range(len(data)):
            data_position.append(int(data[x]))
            
n = max(data_position) - min(data_position)
plt.hist(data_position, bins = n, density = True)
plt.title('Generated position distribution')
plt.xlabel('Position(z)')
plt.ylabel('Probability density')
plt.show()
```

```python
#plot the distribution of generated frequency (away from the center frequency) for Lyman alpha photons
import matplotlib.pyplot as plt

data_frequency = []

#the open function will be modified depend on your own directory
with open("D://Data/Lyf.txt") as f:
    for line in f:
        data = line.split()
        for x in range(len(data)):
            data_frequency.append(float(data[x]))
            
fig, ax = plt.subplots(figsize=(20, 10))
ax.hist(data_frequency, histtype='step', bins=20000000, density=True, linewidth=2)
ax.set_xscale('symlog')
ax.set_xticks([-1e12, -1e9, -1e6, -1e3, 0, 1e3, 1e6, 1e9, 1e12])
ax.tick_params(axis='both', which='both', labelsize=25, width=1.2)
plt.title('Generated frequency distribution', size=25)
plt.xlabel('Frequency(Hz)', size=25)
plt.ylabel('Prbability density', size=25)
```

```python
#plot the distribution of Lyman-alpha propagating in setting environment
import matplotlib.pyplot as plt

data_propagate = []

#the open function will be modified depend on your own directory
with open("D://Data/Lypp.txt") as f:
    for line in f:
        data = line.split()
        for x in range(len(data)):
            data_propagate.append(float(data[x]))
            
n = int((max(data_propagate) - min(data_propagate))/0.000001)
plt.hist(data_propagate, bins = n, density = True)
plt.title('Propagating distance distribution')
plt.xlabel('Propagating distance(dz)')
plt.ylabel('Probability density')
plt.show()
```

```python
#plot the distribution of the status of pre-setted Lyman alpha photons after scattering
import matplotlib.pyplot as plt

data_direction = []
data_polarization = []
data_frequency = []

#the open function will be modified depend on your own directory
with open("D://Data/lys.txt") as f:
    for _ in range(3):
        next(f)
    for line in f:
        data = line.split()
        for x in range(0, len(data), 3):
            data_direction.append(float(data[x]))
            data_polarization.append(float(data[x+1]))
            data_frequency.append(float(data[x+2]))
            
plt.hist(data_direction, bins = 1000, density = True)
plt.title('Scattering direction distribution')
plt.xlabel('Direction(mu)')
plt.ylabel('Probability density')
plt.show()

plt.hist(data_polarization, bins = 1000, density = True)
plt.title('Scattering polarization distribution')
plt.xlabel('Polarization')
plt.ylabel('Probability density')
plt.show()

fig, ax = plt.subplots(figsize=(20, 10))
ax.hist(data_frequency, histtype='step', bins=20000000, density=True, linewidth=2)
ax.set_xscale('symlog')
ax.set_xticks([-1e12, -1e9, -1e6, -1e3, 0, 1e3, 1e6, 1e9, 1e12])
ax.tick_params(axis='both', which='both', labelsize=25, width=1.2)
plt.title('Scattering frequency distribution', size=25)
plt.xlabel('Frequency(Hz)', size=25)
plt.ylabel('Prbability density', size=25)
```
